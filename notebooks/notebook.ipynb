```python
import pandas as pd

df = pd.read_csv("/home/bccestari/Desktop/VM_Share/tools.csv")
```

```python
num_of_records = len(df)
print(num_of_records)
```

```python
df['TOOLS'] = df['TOOLS'].str.split(',')
df = df.explode('TOOLS')
# Reset the index to get consecutive row numbers
df = df.reset_index(drop=True)
```

```python
df
```

```python
df1 = df['TOOLS'].value_counts()
df1 = df1.reset_index()
```

```python
df1.columns = ['TOOLS_Group', 'Count']
df1
```

```python
# Calculate the percentage for each row and add it as a new column
df1['Percentage'] = (df1['Count'] / num_of_records) * 100

df1
```

```python
df1.reset_index(drop=True, inplace=True)
df1.to_csv('/home/bccestari/Desktop/VM_Share/tools_count.csv', index=False)
```
